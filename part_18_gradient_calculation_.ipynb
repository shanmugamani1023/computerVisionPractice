### **Given Values**
- Inputs: \(x_1 = 2\), \(x_2 = 3\)
- True Outputs: \(y_1 = 4\), \(y_2 = 5\)
- Weight: \(w = 1\)
- Bias: \(b = 1\)

### **Step 1: Make Predictions**

The prediction function for a single input is given by:
\[
\hat{y} = w \cdot x + b
\]

#### **Predictions**
1. For \(x_1 = 2\):
   \[
   \hat{y}_1 = w \cdot x_1 + b = 1 \cdot 2 + 1 = 2 + 1 = 3
   \]
2. For \(x_2 = 3\):
   \[
   \hat{y}_2 = w \cdot x_2 + b = 1 \cdot 3 + 1 = 3 + 1 = 4
   \]

### **Step 2: Calculate the Loss**

We will use **Mean Squared Error (MSE)** as our loss function. The formula is:
\[
\text{Loss} = \frac{1}{2} \sum (\hat{y}_i - y_i)^2
\]

#### **Calculating the Loss**
1. For the first prediction:
   \[
   (\hat{y}_1 - y_1)^2 = (3 - 4)^2 = (-1)^2 = 1
   \]
2. For the second prediction:
   \[
   (\hat{y}_2 - y_2)^2 = (4 - 5)^2 = (-1)^2 = 1
   \]

Now we compute the total loss:
\[
\text{Loss} = \frac{1}{2} \left(1 + 1\right) = \frac{1}{2} \cdot 2 = 1
\]

### **Step 3: Calculate the Gradients (Partial Derivatives)**

#### **1. Gradient with respect to \(w\)**
Using the formula for the gradient:
\[
\frac{\partial \text{Loss}}{\partial w} = \sum (\hat{y}_i - y_i) \cdot x_i
\]

Calculating it step-by-step:
- For \(i = 1\):
  \[
  (\hat{y}_1 - y_1) \cdot x_1 = (3 - 4) \cdot 2 = -1 \cdot 2 = -2
  \]
- For \(i = 2\):
  \[
  (\hat{y}_2 - y_2) \cdot x_2 = (4 - 5) \cdot 3 = -1 \cdot 3 = -3
  \]

Now, summing these:
\[
\frac{\partial \text{Loss}}{\partial w} = -2 + (-3) = -5
\]

#### **2. Gradient with respect to \(b\)**
Using the formula:
\[
\frac{\partial \text{Loss}}{\partial b} = \sum (\hat{y}_i - y_i)
\]

Calculating it:
- For \(i = 1\):
  \[
  \hat{y}_1 - y_1 = 3 - 4 = -1
  \]
- For \(i = 2\):
  \[
  \hat{y}_2 - y_2 = 4 - 5 = -1
  \]

Now, summing these:
\[
\frac{\partial \text{Loss}}{\partial b} = -1 + (-1) = -2
\]

### **Step 4: Update the Weights and Bias**

Let’s assume a learning rate \(\eta = 0.1\).

#### **1. Update \(w\)**
Using the update rule:
\[
w = w - \eta \cdot \frac{\partial \text{Loss}}{\partial w}
\]
Substituting the values:
\[
w = 1 - 0.1 \cdot (-5) = 1 + 0.5 = 1.5
\]

#### **2. Update \(b\)**
Using the update rule:
\[
b = b - \eta \cdot \frac{\partial \text{Loss}}{\partial b}
\]
Substituting the values:
\[
b = 1 - 0.1 \cdot (-2) = 1 + 0.2 = 1.2
\]

  \(\text{Loss} = 1\)
